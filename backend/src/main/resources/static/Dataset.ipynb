```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# -----------------------------
# Configuration
# -----------------------------

NUM_LOANS = 500
NUM_DEPOSITS = 500
NUM_TREASURY = 100
NUM_GL = 100

BRANCHES = ["CHN01","CHN02","MUM01","DEL01","BLR01","HYD01","KOL01"]
CURRENCIES = ["INR"]
LOAN_TYPES = ["Home Loan","Personal Loan","Auto Loan","Corporate Loan"]
DEPOSIT_TYPES = ["Savings","Current","Fixed Deposit"]
ACCOUNT_TYPES = ["Asset","Liability","Income","Expense"]

start_date = datetime(2023,1,1)
end_date = datetime(2025,1,1)

# -----------------------------
# Helper Functions
# -----------------------------

def random_date():
    delta = end_date - start_date
    return start_date + timedelta(days=random.randint(0, delta.days))

def maturity_date(start):
    return start + timedelta(days=random.randint(365,3650))

def random_currency():
    return "INR"
# -----------------------------
# Loan Dataset
# -----------------------------

loans = []

for i in range(NUM_LOANS):

    start = random_date()

    loans.append({
        "LoanID": f"LN{i+1}",
        "CustomerID": random.randint(100000,999999),
        "BranchID": random.choice(BRANCHES),
        "LoanType": random.choice(LOAN_TYPES),
        "LoanAmount": round(random.uniform(50000,10000000),2),
        "InterestRate": round(random.uniform(6.5,14),2),
        "Currency": random_currency(),
        "StartDate": start,
        "MaturityDate": maturity_date(start),
        "Status": random.choice(["Active","Closed","Default"])
    })

loan_df = pd.DataFrame(loans)

# -----------------------------
# Deposits Dataset
# -----------------------------

deposits = []

for i in range(NUM_DEPOSITS):

    open_date = random_date()

    deposits.append({
        "DepositID": f"DP{i+1}",
        "CustomerID": random.randint(100000,999999),
        "BranchID": random.choice(BRANCHES),
        "DepositType": random.choice(DEPOSIT_TYPES),
        "Amount": round(random.uniform(1000,5000000),2),
        "InterestRate": round(random.uniform(2,7),2),
        "Currency": random_currency(),
        "OpenDate": open_date,
        "MaturityDate": maturity_date(open_date)
    })

deposit_df = pd.DataFrame(deposits)

# -----------------------------
# Treasury Dataset
# -----------------------------

instruments = [
    "Government Bond",
    "Corporate Bond",
    "FX Forward",
    "Interest Rate Swap",
    "Treasury Bill"
]

treasury = []

for i in range(NUM_TREASURY):

    trade = random_date()

    treasury.append({
        "TradeID": f"TR{i+1}",
        "Instrument": random.choice(instruments),
        "Counterparty": f"CP{random.randint(1000,9999)}",
        "Notional": round(random.uniform(1000000,50000000),2),
        "Currency": random_currency(),
        "TradeDate": trade,
        "MaturityDate": maturity_date(trade)
    })

treasury_df = pd.DataFrame(treasury)

# -----------------------------
# General Ledger Dataset
# -----------------------------

gl_records = []

for i in range(NUM_GL):

    debit = round(random.uniform(1000,500000),2) if random.random() > 0.05 else -1000.0
    credit = round(random.uniform(1000,500000),2) if random.random() > 0.05 else -1000.0

    gl_records.append({
        "GLID": f"GL{i+1}",
        "AccountNumber": random.randint(10000000,99999999),
        "BranchID": random.choice(BRANCHES),
        "AccountType": random.choice(ACCOUNT_TYPES),
        "Debit": debit,
        "Credit": credit,
        "Currency": random_currency(),
        "TransactionDate": random_date()
    })

gl_df = pd.DataFrame(gl_records)

# -----------------------------
# Save CSV Files
# -----------------------------

loan_df.to_csv("loans_dataset.csv", index=False)
deposit_df.to_csv("deposits_dataset.csv", index=False)
treasury_df.to_csv("treasury_dataset.csv", index=False)
gl_df.to_csv("gl_dataset.csv", index=False)

print("Synthetic datasets generated successfully.")
```
